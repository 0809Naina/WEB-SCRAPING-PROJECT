# Introduction To Multiple Pages On A Website

```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page="
```

```python
r = requests.get(url)
```

```python
print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
print(soup)
```

```python
np = soup.find("a",class_ ="_1LKTO3").get("href")
```

```python
print(np)
```

```python
cnp = "https://www.flipkart.com/" + np 
```

```python
print(cnp)
```

```python
Names = []
Price = []
Desc = []
Reviews = []
```

```python
names = soup.find_all("div" , class_ ="_4rR01T")
for i in names:
    n = i.text
    Names.append(n)
    
print(Names)
```

```python
price = soup.find_all("div", class_ = "_30jeq3 _1_WHN1")
for i in price:
    n= i.text
    Price.append(n)
    
print(Price)
```

```python
desc = soup.find_all("div",  class_ = "_3pLy-c row")
for i in desc:
    n = i.text
    Desc.append(n)
    
print(Desc)
```

```python
ratings = soup.find_all("div", class_= "gUuXy-")
for i in ratings:
    n = i.text
    Reviews.append(n)
    
print(Reviews)
```

```python
for i in range(1,11) :
    url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page=" + str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")
    
    names = soup.find_all("div" , class_ ="_4rR01T")
    for i in names:
        n = i.text
        Names.append(n)
    
    print(len(Names))
    
    price = soup.find_all("div", class_ = "_30jeq3 _1_WHN1")
    for i in price:
        n= i.text
        Price.append(n)
    
    print(len(Price))
    
    desc = soup.find_all("div",  class_ = "_3pLy-c row")
    for i in desc:
        n = i.text
        Desc.append(n)
    
    print(len(Desc))
    
    ratings = soup.find_all("div", class_= "gUuXy-")
    for i in ratings:
        n = i.text
        Reviews.append(n)
    
    print(len(Reviews))    
    
```

```python
df = pd.DataFrame({"Product_name":Names, "Product_Description": Desc, "Price_of_Product":Price,"Reviews_of_the_Product":Reviews})
```

```python
print(df)
```

```python
df.to_csv("Mobiles.csv")
```
