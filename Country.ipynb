```python
import re 
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
u = "https://www.scrapethissite.com/pages/simple/"
```

```python
r = requests.get(u)
```

```python
print(r)
```

```python
soup = bs(r.text, "lxml")
soup
```

```python
Noc = []
name = soup.find_all("h3", class_ ="country-name")
for i in name:
    n = i.text.strip()
    Noc.append(n)
print(Noc)
```

```python
Cap =[]
cap = soup.find_all("span", class_="country-capital")
for i in cap:
    n = i.text
    Cap.append(n)
print(Cap)
```

```python
Pop= []
pop = soup.find_all("span",class_ ="country-population")
for i in pop:
    n = i.text
    Pop.append(n)
print(Pop)
```

```python
A = []
area = soup.find_all("span", class_="country-area")
for i in area:
    n = i.text
    A.append(n)
print(A)
```
