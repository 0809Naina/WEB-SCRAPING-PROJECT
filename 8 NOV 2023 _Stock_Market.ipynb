```python
import requests
import pandas as pd
from bs4 import BeautifulSoup 
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
```

```python
r = requests.get(url)
```

```python
print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
print(soup)
```

```python
names = soup.find_all("a", class_ = "title")
```

```python
print(names)
```

```python
product_name = []
```

```python
for i in names :
    name = i.text
    product_name.append(name)
```

```python
print(product_name)
```

```python
prices = soup.find_all("h4", class_ = "float-end price card-title pull-right")
```

```python
prices_list= []
```

```python
for i in prices:
    price = i.text
    prices_list.append(price)
```

```python
print(prices_list)
```

```python
desc =  soup.find_all("p", class_ = "description")
```

```python
desc_lists = []
```

```python
for i in desc:
    desc = i.text
    desc_lists.append(desc)
```

```python
print(desc_lists)
```

```python
reviews = soup.find_all("p",class_="float-end review-count")
```

```python
review_list= []
```

```python
for i in reviews:
    rev = i.text
    review_list.append(rev)
```

```python
print(review_list)
```

```python
print(soup.find("review"))
```

```python
df = pd.DataFrame({"Product Name" : product_name, 
                   "Prices": prices_list,
                   "Description": desc_lists,
                   "Number of Reviews": review_list})
```

```python
print(df)
```

```python
df.to_csv("product_details.csv")
```

# Scrapping A Table From Website -1 

```python
url2="https://ticker.finology.in/"
```

```python
r2 = requests.get(url2)
```

```python
print(r2)
```

```python
soup2 = BeautifulSoup( r2.text , "lxml")
```

```python
soup2
```

```python
soup2.table
```

```python
table = soup2.find("table",class_ ="table table-sm table-hover screenertable")
```

```python
print(table)
```

```python
headers = table.find_all("th")
```

```python
print(headers)
```

```python
titles = []
```

```python
for i in headers:
    x = i.text
    titles.append(x)
```

```python
titles
```

```python
df = pd.DataFrame(columns= titles)
print(df)
```
